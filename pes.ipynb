```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
train
```

```python
#cleaning the data na7ina eli lezmou ytna7 eli hwa el nul
train.replace(' NaN',None,inplace=True)
train.dropna(inplace=True)
train.reset_index(drop=True, inplace=True)
Y = train["Prod"]
X = train.drop(columns=["Prod"])
#X = pd.concat([train["HUMIDITE_RELATIVE(%)"],train["TEMPERATURE(°C)"]],axis=1)
```

```python
# important badelna les valeur mta3 Y bich ywali fi blast ye5id min 1 li 159 7ata lin 1-4
categories = {25: 1, 50: 2, 100: 3, 150: 4}
out = []
for i in Y:
  d = 1000
  for j in categories.keys():
    if i<=j :
      d = min(d,j)
  out.append(categories[d])
Y = pd.Series(out)
```

```python
selector = selector.fit(X,Y)
selector.ranking_
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
#important algorithm mta3 Machine learning "tol 3lih "
clf = DecisionTreeClassifier(random_state=0).fit(X_train,Y_train)
```

```python
prediction = clf.predict(X_test)
score = accuracy_score(Y_test,prediction)
```
