```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns;
sns.set()
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import roc_auc_score, f1_score, confusion_matrix
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("predictive_maintenance.csv")
data.head()
```

```python
data.shape
```

```python
# Drop unnecessary columns
data.drop(['Product ID', 'UDI'], axis=1, inplace=True)
data.head()
```

Counting the classes

```python
data['Target'].value_counts()
```

```python
data['Target'].value_counts().plot.bar(rot=0)
plt.show()
```

```python
#Counting the failure types
data['Failure Type'].value_counts()
```

```python
# Changing the titles of the failures
data['Failure Type'].replace(to_replace=['No Failure', 'Heat Dissipation Failure', 'Power Failure', 'Overstrain Failure', 'Tool Wear Failure', 'Random Failures'],
value=['NoF', 'HDF', 'PF', 'OF', 'TWF', 'RF'], inplace=True)
data.head()
```

```python
data = pd.get_dummies(data)
data.head()
```

```python
data.drop('Failure Type_NoF', axis=1, inplace=True)
tempt = data.pop('Target')
data = pd.concat([data, tempt], 1)
data.head()
```

```python
data.shape
```

```python
X_cols = ['Air temperature [K]', 'Process temperature [K]','Rotational speed [rpm]',
          'Torque [Nm]', 'Tool wear [min]','Type_H','Type_L', 'Type_M']

y_cols = ['Failure Type_OF', 'Failure Type_PF', 'Failure Type_PF', 'Failure Type_TWF', 'Target' ]

X = data[X_cols]
y = data[y_cols]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, shuffle=True, random_state=42)
```

```python
model = RandomForestClassifier(random_state=42)


# Cross Validation and Grid Search
param_grid = {
    'n_estimators': [50, 100, 200, 500],
    'max_features': ['auto', 'log2'],
    'max_depth' : [ 8, 11, 13, 15, 17],
    'criterion' :['gini', 'entropy']
}

CV_model = GridSearchCV(estimator=model, param_grid=param_grid, cv= 5)
CV_model.fit(X_train, y_train)
```

```python
# Getting the best parameters
param = CV_model.best_params_
param
```

```python
# Bulding the model with the best parameters
mdl = RandomForestClassifier(max_depth= param['max_depth'],
                             max_features=param['max_features'],
                             n_estimators=param['n_estimators'],
                             criterion=param['criterion'],
                             random_state=42)
```

```python
mdl.fit(X_train, y_train)
y_pred = mdl.predict(X_test)
```

```python
print('F1 Score is :', "{:.3f}".format(f1_score(y_pred, y_test, average='micro')))
```
